```python
import pandas
import numpy as np
```

```python
class Perceptron:
	# inicia pesos aleatorios
    def __init__(self, n):
        self.weights = np.random.uniform(-1, 1, n)
        self.bias = np.random.uniform(-1, 1)

	# calcula sa√≠da
    def feedFoward(self, x):
        return np.sign(np.dot(self.weights, x) + self.bias)

	#atualiza pesos
	#wn = wa + (t-y) * x
    def updateWeights(self, x, error):
        self.weights += error * x
        self.bias += error

	# treinamento com X exemplos e seus T targets
    def train(self, X, T, epochs= 1000):
        for _ in range(epochs):
            sumError = 0

            for x, t in zip(X, T):
                error = t - self.feedFoward(x)
                self.updateWeights(x, error)
                sumError += np.abs(error)
                print(error)
            print(error)

			# se estiver certo antes das mil epocas, break
            if sumError == 0.0:
                break
```

```python
df = pandas.read_csv("linsep.csv", names=['x1', 'x2', 'y'])

df.head()
```

```python
inputs = df[['x1', 'x2']].values
outputs = df['y'].values
```

```python
net = Perceptron(2)
net.train(inputs, outputs)
```
